## Activation functions in Neural networks

Activation function is a mathematical function applied to a neuron's output to introduce non-linearity into the model, enabling it to learn complex patterns and relationships in the data.

It determines whether a neuron should be activated or not by transforming the input signal into an output signal, often bounded within a specific range.

Without activation functions, neural networks would behave as linear models, regardless of their depth, and fail to solve non-linear problems.

They play an important role in guiding the learning process by affecting gradient propagation and the model's ability to converge effectively during training. 

Different activation functions, such as ReLU, sigmoid, tanh, and softmax, are chosen based on the specific requirements of the task and network architecture.

There are several activation functions used in deep learning.

1. Sigmoid activation function
2. Tanh activation function
3. ReLU (Rectified Linear Unit) function
4. Softmax function

### 1. Sigmoid activation function

One of the popular activation functions used in deep learning.

It is commonly used in output layer of binary classification tasks

The output of the sigmoid functions ranges between 0 and 1. So, it can predict probabilities or binary outcomes

Sigmoid function is non-linear allowing the network to capture complex patterns

It is a smooth and differentiable function ,so it is suitable for gradient-based optimization methods like backpropogation

It is a saturating function. For very large input values, the ouput approaches 1 and for very small input values, the output approaches 0. This can lead to vanishing gradient problems, slowing down learning in deep networks.

The ouput is always positive, which can lead to inefficienies during training since the gradients will also always be positive or negative


Formula for sigmoid activation function

$$\sigma(x) = \frac{1}{1 + e^{-x}}$$


Where, x is the input given to the neuron

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1.0/(1.0+ np.exp(-x))

x = np.linspace(-10,10,100)
y = sigmoid(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('Ïƒ(x)')
plt.title('Sigmoid Activation Function')
plt.show()
```

### 2. Tanh activation function

The hyperbolic tangent activation function is another widely used activation function in neural networks.

The output of the tanh function ranges between -1 and 1.

It is a non-linear function, allowing the network to model complex relationships.

It is a smooth and differentiable function, making it suitable for gradient-based optimization methods like backpropagation.

Unlike the sigmoid function, tanh is zero-centered. The output is symmetrically distributed around zero. This helps in balancing the gradients and thereby making learning more efficient.

Tanh can also saturate. For very large or very small input values, the output approaches 1 or -1. This can lead to the vanishing gradient problem, similar to the sigmoid function.

This function is often used in hidden layers of neural networks.

Formula for tanh activation function

$$\text{tanh}(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}$$

Where, x is the input given to the neuron

```python
def tanh(x):
    return np.tanh(x)

x = np.linspace(-10, 10, 100)
y = tanh(x)

plt.plot(x,y)
plt.title("Hyperbolic Tangent Activation Function")
plt.xlabel("x")
plt.ylabel("tanh(x)")
plt.show()
```

### 3. ReLU (Rectified Linear Unit) function

The ReLU (Rectified Linear Unit) activation function is one of the most widely used activation functions in modern neural networks

The output of the ReLU function ranges from 0 to infinity. The function outputs zero for any negative input and the input itself for any positive input.

ReLU is non-linear, allowing the network to model complex patterns and interactions between features.

Unlike sigmoid and tanh, ReLU does not saturate for positive inputs, which helps in avoiding the vanishing gradient problem. This allows for faster learning.

The output of ReLU is not zero-centered since it only outputs non-negative values.

ReLU leads to sparse activation. In a large network, many neurons will output zero for a given input, which can help with computational efficiency and reducing overfitting.

Main drawback of ReLU is the "dying ReLU" problem, where neurons can get stuck outputting zero and stop learning. This happens if the weights are updated in such a way that they always produce a negative input for certain neurons.

Variants like Leaky ReLU, Parametric ReLU (PReLU), and ELU have been introduced to address the dying ReLU problem by allowing a small, non-zero gradient for negative inputs.

Formula for tanh activation function

$$\text{ReLU}(x) = \max(0, x)$$

Where, x is the input given to the neuron

```python
def relu(x):
    return np.maximum(0,x)

x = np.linspace(-10, 10, 100)
y = relu(x)

plt.plot(x, y)
plt.title("ReLU Activation Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

### 4. Softmax function

Softmax function is used in the output layer for multi-class classification tasks.

The output of the softmax function ranges from 0 to 1.

It is a non-linear function.

It transforms raw scores into probabilities, ensuring that the sum of probabilities across classes is 1.

Softmax function turns logits value into probabilities by taking the exponents of each output and then normalizing each number by the sum of those exponents so that the entire output vector adds up to one (Logits are the raw score values produced by the last layer of the neural network before applying any activation function on it)

Formula for softmax function

$$\text{softmax}(x_i) = \frac{e^{x_i}}{\sum_{j=1}^{n} e^{x_j}}$$
